```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import nltk
import regex as re
```

```python
fake = pd.read_csv('Fake.csv')
true = pd.read_csv('True.csv')

fake = fake.drop(columns = ['title','date', 'subject'])
true = true.drop(columns = ['title','date', 'subject'])

fake['real'] = 0
true['real'] = 1

news = pd.concat([fake,true], ignore_index = True).sample(frac = 1).reset_index(drop = True)

news2 = news

stopeng = stopwords.words('english')

stop_regex = r'\b(?:{})\b'.format('|'.join(stopeng))

print(news.text[0])

news.text = news.text.str.lower().str.replace(stop_regex, '', regex = True).str.replace(r'[^a-z]+', ' ', regex = True).str.split(r'\s', regex = True)
news.text.apply(lambda words: [word for word in words if word])


# news.text = [w.replace(stop_regex, '') for w in news.text[][:-1]]
```
